## **INTRODUCTION**
Per capita income is a key economic indicator used to assess the average income level and living standards within a country. This study analyzes long-term per capita income trends in Nigeria and four other African countries using GDP and population data from the World Bank.

---
### **OBJECTIVES**
- To compute per capita income using GDP and population data
- To examine income trends over a 30-year period
- To compare Nigeria's performance with other African economies

---
### **KEY FINDINGS**
- South Africa consistently records the **highest per capita income**
- Nigeria shows **slow and unstable growth**, largely due to rapid population increase
- Ghana and Kenya demonstrate steady **improvement over time**
- Egypt maintains relatively strong per capital income due to controlled population growth

---
#### **CONCLUSION**
The analysis reveals significant disparities in per capita income among African countries. While some economies show consistent improvement, Nigeria's income growth remains constrained by population pressure and economic volatility. Sustainable economic policies and population management are critical for improving living standards.

---
#### **RECOMMENDATION**
To strengthen per capita income growth and improve living standards, particularly in Nigeria and similar African economies, the following policy measures are recommended:
- **Economic Diversification**    
Governments should reduce overreliance on a single sector (such as oil in Nigeria) by promoting diversification into agriculture, manufacturing, technology, and services. A broader economic base enhances income stability and increases per capita income growth.
- **Population Management and Human Capital Development**    
Rapid population growth dilutes income gains. Policies promoting access to education, family planning, and healthcare can help stabilize population growth while improving labor productivity and long-term income levels.
- **Investment in Education and skills**  
Increased public and private investment in education, vocational training, and digital skills is essential. A skilled workforce enhances productivity, attracts foreign direct investment, and supports higher income per capita.
- **Infrastructure Development**  
Improving infrastructure, particularly electricity, transportation, water resources, and digital connectivity, reduces production costs and boosts economic efficiency, leading to higher incomes across sectors.
- **Support for Small and Medium Enterprises (SMEs)**  
SMEs are major employers in African economies. Access to affordable credit, business training, and market linkages can significantly improve income generation and per capita income growth.
- **Strengthening Governance and Macroeconomic Stability**  
Sound fiscal management, anti-corruption measures, and stable macroeconomic policies increase investor confidence and ensure sustainable economic growth that translates into higher per capita income.

##### **Concept Used**
Since the GDP dataset is GDP(constant US$) and population is provided:
## **Per Capita Income** = GDP(constant US$)/Total Population

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading datasets
population = pd. read_excel(r"\Users\DELL\Desktop\API_SP.POP.TOTL_DS2_en_excel_v2_2834.xls",
                            sheet_name="Data",
                            skiprows=3)
population
```

```python
gdp = pd.read_excel(r"\Users\DELL\Desktop\API_NY.GDP.MKTP.KD_DS2_en_excel_v2_4197.xls",
                    sheet_name="Data",
                    skiprows=3)
gdp
```

```python
# Fetching out countries of interest
countries = ["Nigeria",
             "Ghana",
             "Kenya",
             "South Africa",
             "Egypt"]
```

```python
# Filtering countries of interest from population dataset
population = population[population["Country Name"].isin(countries)]
population
```

```python
# Filtering countries of interest from GDP dataset
gdp = gdp[gdp["Country Name"].isin(countries)]
gdp
```

```python
# Reshaping the population data
pop_long = population.melt(id_vars=["Country Name"],
                           var_name="Year",
                           value_name="Population")
pop_long
```

```python
# Reshaping the GDP data
gdp_long = gdp.melt(id_vars=["Country Name"],
                    var_name="Year",
                    value_name="GDP")
gdp_long
```

```python
# Converting population to numeric
pop_long["Population"] = pd.to_numeric(pop_long["Population"],
errors="coerce")
pop_long
```

```python
# Converting GDP to numeric
gdp_long["GDP"] = pd.to_numeric(gdp_long["GDP"],
errors="coerce")
gdp_long
```

```python
# Merging datasets
data = pd.merge(pop_long, gdp_long, on=["Country Name", "Year"])
data
```

```python
# Compute per capita income
data["Per Capita Income (USD)"] = data["GDP"]/data["Population"]
data
```

```python
# Filtering last 30 years
data = data[data["Year"] >= data["Year"].max() - 30]
data
```

## **Plotting of Per Capita Income (USD) Trends Graph**

```python
plt.figure(figsize=(16,9))
for country in countries:
    subset = data[data["Country Name"] == country]
    plt.plot(subset["Year"], subset["Per Capita Income (USD)"], label=country)
plt.title("Per Capita Income Trends in Selected African Countries (1993-2023)")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (USD)")
plt.legend(title="Country")
plt.grid(True)
plt.show()
```
